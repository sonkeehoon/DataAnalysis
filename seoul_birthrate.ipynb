```python
# 결측치"-"를 0으로 대체
df_drop = df_drop.replace("-", 0)
```

```python
# 모든 수치데이터를 int나 float으로 변경
df_drop.iloc[:,1:] = df_drop.iloc[:,1:].apply(pd.to_numeric)
```

```python
df_drop
```

```python
# column을 20대,30대,40대로 합치는 작업
df_drop.rename(columns={'2022.2':'15-20세'}, inplace=True)
df_drop['20대'] = df_drop['2022.3']+df_drop['2022.4']
df_drop['30대'] = df_drop['2022.5']+df_drop['2022.6']
df_drop['40대'] = df_drop['2022.7']+df_drop['2022.8']
df_drop.iloc[:, 4:10]
# df_drop = df_drop.drop(df_drop.columns[4:10], axis = 1)
df_drop
```

```python
# 불필요한 column 제거
df_drop.iloc[:, 4:10]
# df_drop = df_drop.drop(df_drop.columns[4:10], axis = 1) drop은 한번만
df_drop
```

```python
# column 이름 변경
df_drop.rename(columns={'자치구별(1)':'자치구별', '2022':'평균 출산 연령', '2022.1':'합계출산'}, inplace=True)
```

```python
plt.figure(figsize=(15,8))
ax = sns.barplot(data=df_drop, x='자치구별', y='20대')
plt.title("서울 20대(2022)")
plt.xlabel('자치구별')
plt.ylabel('출생아 수')
ax.plot(15, 366, "*", markersize=10, color = "r")
plt.show()
```

```python
plt.figure(figsize=(15,8))
ax = sns.barplot(data=df_drop, x='자치구별', y='30대')
plt.title("서울 30대(2022)")
plt.xlabel('자치구별')
plt.ylabel('출생아 수')
ax.plot(23, 2771, "*", markersize=10, color = "r")
plt.show()
```

```python
plt.figure(figsize=(15,8))
ax = sns.barplot(data=df_drop, x='자치구별', y='40대')
plt.title("서울 40대(2022)")
plt.xlabel('자치구별')
plt.ylabel('출생아 수')
ax.plot(23, 232, "*", markersize=10, color = "r")
plt.show()
```
